```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
import pickle

def detector():
    class InputForm(tk.Tk):
        def __init__(self):
            super().__init__()

            self.title("Crop Yield Prediction")
            self.geometry("400x600")

            self.user_inputs_df = pd.DataFrame(columns=[
                "experience", "test_score", "interview_score2", "interview_score3",
                "interview_score4", "interview_score9", "temperature", "wind_speed",
                "precipitation", "humidity", "N", "P", "K", "pressure"
            ])

            self.create_widgets()

        def create_widgets(self):
                    # Labels
                    ttk.Label(self, text="State name").grid(row=0, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="District name").grid(row=1, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Season Type").grid(row=2, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Crop").grid(row=3, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Area").grid(row=4, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Soil Type").grid(row=5, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Temperature").grid(row=6, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Wind Speed").grid(row=7, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Precipitation").grid(row=8, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Humidity").grid(row=9, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="N").grid(row=10, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="P").grid(row=11, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="K").grid(row=12, column=0, padx=10, pady=10, sticky="w")
                    ttk.Label(self, text="Pressure").grid(row=13, column=0, padx=10, pady=10, sticky="w")

                    # Entry Widgets
                    self.state_entry = ttk.Entry(self, width=30)
                    self.district_entry = ttk.Entry(self, width=30)
                    self.season_type_combobox = ttk.Combobox(self, values=["Autumn", "Kharif", "Rabi", "Summer", "Whole Year", "Winter"], width=27)
                    self.crop_entry = ttk.Entry(self, width=30)
                    self.area_entry = ttk.Entry(self, width=30)
                    self.soil_type_combobox = ttk.Combobox(self, values=["Chalky", "Clay", "Loamy", "Peaty", "Sandy", "Silt", "Silty"], width=27)
                    self.temperature_entry = ttk.Entry(self, width=30)
                    self.wind_speed_entry = ttk.Entry(self, width=30)
                    self.precipitation_entry = ttk.Entry(self, width=30)
                    self.humidity_entry = ttk.Entry(self, width=30)
                    self.n_entry = ttk.Entry(self, width=30)
                    self.p_entry = ttk.Entry(self, width=30)
                    self.k_entry = ttk.Entry(self, width=30)
                    self.pressure_entry = ttk.Entry(self, width=30)

                    # Button
                    submit_button = ttk.Button(self, text="Click for the Yield", command=self.submit_form)

                    # Layout
                    self.state_entry.grid(row=0, column=1, padx=10, pady=10)
                    self.district_entry.grid(row=1, column=1, padx=10, pady=10)
                    self.season_type_combobox.grid(row=2, column=1, padx=10, pady=10)
                    self.crop_entry.grid(row=3, column=1, padx=10, pady=10)
                    self.area_entry.grid(row=4, column=1, padx=10, pady=10)
                    self.soil_type_combobox.grid(row=5, column=1, padx=10, pady=10)
                    self.temperature_entry.grid(row=6, column=1, padx=10, pady=10)
                    self.wind_speed_entry.grid(row=7, column=1, padx=10, pady=10)
                    self.precipitation_entry.grid(row=8, column=1, padx=10, pady=10)
                    self.humidity_entry.grid(row=9, column=1, padx=10, pady=10)
                    self.n_entry.grid(row=10, column=1, padx=10, pady=10)
                    self.p_entry.grid(row=11, column=1, padx=10, pady=10)
                    self.k_entry.grid(row=12, column=1, padx=10, pady=10)
                    self.pressure_entry.grid(row=13, column=1, padx=10, pady=10)

                    submit_button.grid(row=14, column=0, columnspan=2, pady=20)

            def submit_form(self):
                # Collecting user input
                input_data = {
                    "experience": self.state_entry.get(),
                    "test_score": self.district_entry.get(),
                    "interview_score2": self.season_type_combobox.get(),
                    "interview_score3": self.crop_entry.get(),
                    "interview_score4": self.area_entry.get(),
                    "interview_score9": self.soil_type_combobox.get(),
                    "temperature": self.temperature_entry.get(),
                    "wind_speed": self.wind_speed_entry.get(),
                    "precipitation": self.precipitation_entry.get(),
                    "humidity": self.humidity_entry.get(),
                    "N": self.n_entry.get(),
                    "P": self.p_entry.get(),
                    "K": self.k_entry.get(),
                    "pressure": self.pressure_entry.get(),
                }

                # Append user input to DataFrame
                self.user_inputs_df = self.user_inputs_df.append(input_data, ignore_index=True)

                # Clear entry fields
                self.clear_entries()

        def clear_entries(self):
            # Clear all entry fields
            for entry in self.winfo_children():
                if isinstance(entry, ttk.Entry) or isinstance(entry, ttk.Combobox):
                    entry.delete(0, 'end')

        def save_to_pickle(self):
            # Save the DataFrame to a pickle file
            with open('user_inputs_df.pkl', 'wb') as f:
                pickle.dump(self.user_inputs_df, f)
            print("User inputs saved to 'user_inputs_df.pkl'")

            # Close the Tkinter app
            self.destroy()

    # Create and run the Tkinter app
    app = InputForm()

    # Button to save inputs and close the app
    save_button = ttk.Button(app, text="Save and Close", command=app.save_to_pickle)
    save_button.pack(pady=20)

    app.mainloop()

# Call the detector function
detector()
```
